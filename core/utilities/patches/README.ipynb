# Patches
This package provides an easy to use API to create custom Patches

### Getting started
#### Empty Patch
The `Patch` class creates an empty patch with a given shape.

```python
%matplotlib notebook
```

```python
empty_p = Patch(shape = (100,100))
empty_p.plot2d()
empty_p.plot3d()
```

#### Create a custom patch
To create a custom patch you have to subclass `Patches` and override the `make` method.  Below there is a simple example for a patch with square in the middle

```python
from patches import *

class SquareInTheMiddlePatch(Patch):
    def make(self, size=4, *args, **kwargs):
#         here we can access the .hm filled with zeros
        self.hm[self.shape[0] // 2 - size : self.shape[0] // 2 + size, self.shape[1] // 2 - size: self.shape[0] // 2 + size] = 0.5
        return self.hm
```

Then always remember to call the class to invoke the `.make` method.

```python
empty_p = SquareInTheMiddlePatch(shape = (100,100))
empty_p(size=24)
empty_p.plot2d()
empty_p.plot3d()
```

#### Existing Patches
Below there are some examples from the already existing patches

```python
WallPatch((88,88))().plot3d()
WallPatch((88,88), strength = 0.2)().plot3d()
BumpsPatch((88,88))().plot3d()
HolesPatch((88,88))().plot3d()
RampPatch((88,88))().plot3d()
RampPatch((88,88), orientation=1)().plot3d()
```
